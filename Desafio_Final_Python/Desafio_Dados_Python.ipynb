```python
import matplotlib.pyplot as plt

# Contar funcionários por setor
department_counts = df['Department'].value_counts()

# Plotar gráfico de barras
plt.figure(figsize=(8, 5))
department_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.xlabel('Departamento')
plt.ylabel('Quantidade de Funcionários')
plt.title('Quantidade de Funcionários por Departamento')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Esse gráfico mostra que a maior quantidade de funcionários trabalha no departamanto 'Research & Development', enquanto o departamento 'Human Resources' concentra a menor quantidade.

```python
# Contar funcionários por nível de escolaridade
education_counts = df['Education'].value_counts().sort_index()

# Plotar gráfico de linha
plt.figure(figsize=(8, 5))
education_counts.plot(kind='line', marker='o', color='red', linestyle='-')
plt.xlabel('Nível de Escolaridade')
plt.ylabel('Quantidade de Funcionários')
plt.title('Quantidade de Funcionários por Nível de Escolaridade')
plt.xticks(education_counts.index)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
```

Esse gráfico mostra que maior parte dos funcionários estão no nível 3 de escolaridade, enquanto o nível 5 concentra a menor quantidade de funcionários.

```python
# Contar funcionários por status de atrito (sim ou não)
attrition_counts = df['Attrition'].value_counts()

# Plotar gráfico de pizza
plt.figure(figsize=(6, 6))
attrition_counts.plot(kind='pie', autopct='%1.1f%%', colors=['pink', 'lightgreen'], startangle=90)
plt.title('Distribuição de Funcionários com ou sem Atrito')
plt.ylabel('')  # Remove rótulo do eixo Y
plt.show()
```

Esse gráfico mostra que a maior parte dos funcionários não têm atrito com a empresa.
